```python
import os
import pandas as pd
from docx import Document
import numpy as np

# Load the Excel file
excel_path = './data/工作簿2.xlsx'
df = pd.read_excel(excel_path)

# # Check the structure of the DataFrame
# print(df.head())

# Filter out rows with NaN values in the required columns
df = df.dropna(subset=['Subtitle', 'Description'])

# Create a new Word document
doc = Document()

# Define a function to add sections to the Word document
def add_section(doc, section_number, section_title, content):
    doc.add_heading(f"{section_number} {section_title}", level=1)
    sub_section_number = 1
    for idx, row in content.iterrows():
        subtitle = row['Subtitle']
        description = row['Description']
        # Check if the subtitle is not NaN
        if pd.notna(subtitle):
            doc.add_heading(f"{section_number}.{sub_section_number} {subtitle}", level=2)
            sub_section_number += 1
        # Check if the description is not NaN
        if pd.notna(description):
            doc.add_paragraph(description)

# Iterate through the Excel data and populate the Word document
section_title = "同步涉敏主账号"
section_number = "4"

# Assuming the Excel file has columns 'Subtitle' and 'Description'
add_section(doc, section_number, section_title, df)

# Save the Word document
word_path = './data/Generated_Document2.docx'
doc.save(word_path)
print(f"Word document saved at {word_path}")
```
